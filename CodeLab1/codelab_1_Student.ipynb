# Part E: Train-Test Split

**E1)** 
Create a training-test split function called “My_split”. The function shuffles the input data and selects random samples to generate training and test data sets determined by another user input “ratio” between [0.0-1.0]. Provide your function in the report, you can provide a clear image of your function as well.  (Note: each row corresponds to a different sample where each column is a separate feature. This is a common notation for the future CodeLabs as well.) 

Divide the data frame of “personAclick” into training and test sets with a ratio of 0.8 (80% Training, 20% Test) where each row represents a single data entry. What are the dimensions of the train and test data frames? Why do we need this process for machine learning applications?


```python
# Function
def My_split(data, ratio):
    # if ratio not in the bound then give error 
    if ratio < 0.0 or ratio > 1.0:
        raise ValueError("Ratio must be between 0.0 and 1.0")

    # shuffle the data
    shuffled_data = data.sample(frac=1).reset_index(drop=True)

    # determine the split index
    split_index = int(ratio * len(shuffled_data))
    
    # split the data according to the split index   
    train = shuffled_data[:split_index]
    test = shuffled_data[split_index:]

    return train, test
```

```python
click=np.load('PersonAclick.npy')
click_df = pd.DataFrame(click)
train,test = My_split(click_df,0.8)
```

# Bonus: Train-Test split of labels

Add a column (label vector) containing 0s and 1s to the data frame of “personAclick” of 800 elements with randomly placed 40-60 split of 0s and 1s (40% 0s, 60% 1s). Split the data frame into training and test sets with a ratio of 0.8, where the ratio of 0s to 1s remains the same. This split process is called a stratified train-test split. Show the ratio of 0s to 1s for both the training and test set.

```python
click=np.load('PersonAclick.npy')
click_df = pd.DataFrame(click)
num_elements = len(click_df)
print(num_elements)
num_zeros = int(0.40 * num_elements)  # 40% 0s
num_ones = num_elements - num_zeros   # 60% 1s
print(num_zeros)
print(num_ones)
labels = np.array([0] * num_zeros + [1] * num_ones) # the label vector 
print(labels)
np.random.shuffle(labels) # shuffle the label vector
print(labels)
```

```python
click_df['label'] = labels # the label vector is added as the last column
print(click_df)
```

```python
train_data, test_data = My_split(click_df, 0.8)
```

```python
train_ratio = train_data['label'].value_counts(normalize=True) # gives frequencis
test_ratio = test_data['label'].value_counts(normalize=True) # gives frequencis

print("Training set ratio of 0s to 1s:\n", train_ratio)
print("Test set ratio of 0s to 1s:\n", test_ratio)
```
